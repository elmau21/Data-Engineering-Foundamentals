# Meanings:

- Pivot: Convert rows to columns.
- Melt: Convert columns to rows.
- Stack: Convert columns to rows (multi-index).
- Unstack: Convert rows to columns (multi-index).

```python
#orderBy/groupBy/sort(ascending=True/False, descensing=True/False)

# Linear Search doesn't need the array to be sorted.

#Note: I usually write bad data/Corrupted data to make y'all think about solutions.

data_new = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-03', '2023-01-03'], #The dates was in random order, which represents a problem with dealing with the correct format
    'Product': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Quantity': [10, 15, 12, 9, 8, 6]
}

df = pd.DataFrame(data_new)
df['Quantity'] = df['Quantity'].astype(float) #Change Quantity values from INT to Float
df.head(10)
```

Pivoting a Dataset
- Parameters for pivot_table:
    - values: Column(s) to aggregate.
    - index: Column(s) for row index.
    - columns: Column for columns in pivot.
    - aggfunc: Aggregation function (default: 'mean').
    - sfill_value: Replace missing values.
    - margins: Add row/column subtotals (default: False).
    - dropna: Exclude rows with NaN (default: True).

```python
pivot_df = df.pivot_table(index='Date', columns='Product', values='Quantity', aggfunc=sum) #.sum()
pivot_df.head()
```

Melting a Dataset
- .melt(): Unpivot DataFrame, transform columns into rows.
    - Parameters:

    - id_vars: Column(s) to keep as identifier variables.
    - value_vars: Column(s) to unpivot (melt).

```python
melted_df = df.melt(id_vars='Date', value_vars='Product')
melted_df.head()
```

# Stacking a DataSet
- .stack(): Pivot DataFrame, moves column index to row index.
- .reset_index(): Resets index, new DataFrame with 'Quantity' column.
  - Parameters:
    - name='Quantity': Name of the new column containing stacked values.

```python
stacked_df = df.set_index(['Date','Product']).stack().reset_index(name='Quantity')
stacked_df.head()
```

# Unstacked a DataSet
- .unstack(): Pivot DataFrame, moves row index to column index.

```python
#Hint: Is there anything to unstack in the original df?
#Hint2: Did we apply something before?
#Hin3: If the melting method is the opposite of the pivot method and unstack is the opposite of the stack, who needs to be unstacked?
#Hint4: which df?

unstacked_df = stacked_df.set_index(['Date', 'Product']).unstack()
unstacked_df.head()
```

## Understanding a boxplot:

- Box: The box represents the interquartile range (IQR), which is the middle 50% of the data. It spans from the 25th percentile (Q1) to the 75th percentile (Q3).

- Line inside the box: This line represents the median, which is the middle value of the dataset when it is ordered from the smallest to the largest. Half of the data points are below this line, and half are above it.

- Whiskers: These are the lines that extend from the box. They represent the range of the data within a certain factor (usually 1.5 times the IQR) from the upper and lower quartiles. Data points beyond the whiskers are considered outliers.

- Outliers: Individual data points that lie beyond the whiskers are plotted as individual points. Outliers are usually shown as circles or dots.

# NOTE:

In a box plot, the green line represents the "median" of the data. The box plot is a graphical representation of the distribution of the data, and it provides a visual summary of key statistics.

# Concat method

To combine (concatenate) two or more DataFrames along a certain axis, use pandas' concat() function. You can arrange DataFrames either horizontally along columns or vertically along rows. All of the input DataFrames' combined data will be present in the final DataFrame.

- Parameters:
  - objs: This is a required parameter and refers to a list or sequence of DataFrames that you want to concatenate. It can also be a dictionary of DataFrames.

  - axis: This parameter specifies the axis along which the concatenation should be done. It can take one of two values: 0 (default) for vertical concatenation (along rows), and 1 for horizontal concatenation (along columns).

  - join: This parameter determines how the columns from different DataFrames are combined. The options are 'outer' (default) for union of all columns, and 'inner' for intersection of columns.

  - ignore_index: If set to True, the resulting DataFrame will have a new index from 0 to N-1, where N is the total number of rows. If set to False (default), the index from the original DataFrames will be retained.

# Applying Functions and transformations to data
  - Apply
  - map

```python
raw_data = {
    'Name': ['Mariano', 'Alexis', 'Javi', 'Willy'],
    'Salary': [5000, 7000, 6000, 5500],
    'Deparment': ['HR', 'IT', 'Marketing', 'Finance']
}

df6 = pd.DataFrame(raw_data)
df6
```

```python
#Function to calculate bonus

def bonus(salary):
  if salary > 6000:
    return salary * 0.1
  else:
    return salary * 0.05

def check_is_rich(salary):
  if salary > 6000:
    return 'Rich as fuck'
  else:
    return 'Kinda Broke'

#Apply the function to create a new column
df6['Bonus'] = df6['Salary'].apply(bonus)
df6['Status'] = df6['Salary'].apply(check_is_rich)
df6
```

# .map() method is a python func that allows you to transform elements in an interable data structure, based on a mapping provided by a dict, a funct or a SERIES

```python
#Mapping departments to abbreviations

department_map = {'HR': 'Human Resources', 'Finance': 'FIN', 'IT': 'Information Technology', 'Marketing': 'Mktg'}

#Using .map() function to replace values
df6['Department Abreviation'] = df6['Deparment'].map(department_map)
df6
```
